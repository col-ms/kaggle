```python
# Import packages
import numpy as np
import pandas as pd

import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
# Loading both personnel and equipment loss data
data_eq = pd.read_csv('../input/2022-ukraine-russian-war/russia_losses_equipment.csv')
data_per = pd.read_csv('../input/2022-ukraine-russian-war/russia_losses_personnel.csv')
```

Please note that all following graphs show cumulative records.

**Personnel Losses**

```python
# Simple Personnel over time graph
fig = go.Figure()
fig.add_trace(go.Scatter(x = data_per['date'],
                         y = data_per['personnel'],
                         mode = 'lines+markers'))
fig.update_layout(
    title = 'Russian Personnel Losses',
    xaxis_title = 'Date',
    yaxis_title = 'No. of Personnel Lost')
fig.show()
```

**Equipment Losses**

```python
# Create date variable as it will be used in each trace
date = data_eq['date']

# Create plot
fig = go.Figure()
fig.add_trace(go.Scatter(x = date, y = data_eq['aircraft'],
                         mode = 'lines+markers',
                         name = 'aircraft'))
fig.add_trace(go.Scatter(x = date, y = data_eq['helicopter'],
                         mode = 'lines+markers',
                         name = 'helicopter'))
fig.add_trace(go.Scatter(x = date, y = data_eq['anti-aircraft warfare'],
                         mode = 'lines+markers',
                         name = 'anti-aircraft warfare'))
fig.add_trace(go.Scatter(x = date, y = data_eq['drone'],
                         mode = 'lines+markers',
                         name = 'drone'))
fig.update_layout(legend_orientation = 'h',
                  legend = dict(x = 0, y = 1, traceorder = 'normal'),
                  title = 'Equipment Losses: Air',
                  yaxis_title = 'No. of Losses',
                  xaxis_title = 'Date')
fig.show()
```

```python
# Again, create variable to make referencing easier
date = data_eq['date']

# Create plot
fig = go.Figure()
fig.add_trace(go.Scatter(x = date, y = data_eq['tank'],
                         mode = 'lines+markers',
                         name = 'tank'))
fig.add_trace(go.Scatter(x = date, y = data_eq['field artillery'],
                         mode = 'lines+markers',
                         name = 'field artillery'))
fig.add_trace(go.Scatter(x = date, y = data_eq['APC'],
                         mode = 'lines+markers',
                         name = 'APC'))
fig.add_trace(go.Scatter(x = date, y = data_eq['military auto'],
                         mode = 'lines+markers',
                         name = 'military auto'))
fig.update_layout(legend_orientation = 'h',
                  legend = dict(x = 0, y = 1, traceorder = 'normal'),
                  title = 'Equipment Losses: Ground/Other',
                  yaxis_title = 'No. of Losses',
                  xaxis_title = 'Date')
fig.show()
```
